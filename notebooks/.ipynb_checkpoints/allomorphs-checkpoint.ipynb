```python
def list_from_letters(word):
    letter_list = []
    n_for_list = 0
    for n, letter in enumerate(word):
        if letter == "'" or letter == "̥":
            if word[n-1] == "'" or word[n-1] == "̥":
                letter_list[n_for_list - 1] = word[n-2:n+1]
            else:
                letter_list[n_for_list - 1] = word[n-1:n+1]                 
        else:
            letter_list.append(letter)
            n_for_list += 1 
    n = 1
    for letter in letter_list[1:]:
        if letter == letter_list[n-1]:
            letter_list[n-1] = letter + letter
            del letter_list[n]
        else:
            n += 1
    n = 1
    for letter in letter_list[1:]:
        if letter == 'a' or letter == 'e':
            if letter_list[n-1] == 'o' or letter_list[n-1] == 'u':
                letter_list[n-1] += letter
                del letter_list[n]
            else:
                n += 1
        else:
                n += 1
    return letter_list
```

### Функция для создания парадигм

```python
with open('../intremediate files/classes_noun_transcript.json', encoding='utf-8') as f:
    verb_alt_transcript_dict = json.load(f)
```

```python
sg = ['NOM.SG', 'ACC.SG', 'GEN.SG', 'ILL.SG', 'LOC.SG', 'COM.SG', 'ABE.SG']
pl = ['NOM.PL', 'ACC.PL', 'GEN.PL', 'ILL.PL', 'LOC.PL', 'COM.PL', 'ABE.PL']
other = ['ESS', 'PART']

inflections = {'LOC.SG': ["es't'","s'"], 
              'COM.SG': ["en'"], 
              'ABE.SG': ['xa','xta'], 
              'ACC.PL': ["et'"], 
              'ILL.PL': ["et'"], 
              'LOC.PL': ["en'"], 
              'COM.PL': ["eguim","goim","geim"], 
              'ABE.PL': ["exa", "xta"], 
              'ESS': ["en'"], 
              'PART': ["dde"]}


def noun_paradigm(noun, n_class, n_type=0):  
    paradigm = dict.fromkeys(['NOM.SG', 'ACC.SG', 'GEN.SG', 'ILL.SG', 'LOC.SG', 'COM.SG', 'ABE.SG', 'NOM.PL', 'ACC.PL', 'GEN.PL', 'ILL.PL', 'LOC.PL', 'COM.PL', 'ABE.PL', 'ESS', 'PART'])
    letters = list_from_letters(noun)
    if n_class == 1:
        # на долгий согласный СС
        strong = noun
        weak = noun[:-len(letters[-1])//2]
        
        for case in paradigm:
            # чередуем гласные
            for alt_cases in noun_alt_transcript_dict[str(n_type)]:
                if case in alt_cases['cases']:
                    found_alt = ''
                    for search_alt in alt_cases['alt']:
                        if re.search(search_alt, noun) != None:
                            found_alt = search_alt
                    if found_alt != '':
                        for sub_alt in alt_cases['alt'][found_alt]:
                            noun_alt = re.sub(found_alt, sub_alt, noun, count=1)
                            letters = list_from_letters(noun_alt)
                        
            strong = noun_alt
            weak = noun_alt[:-len(letters[-1])//2]
            # изменяем основу
            if case in pl + ['ACC.SG', 'GEN.SG', 'LOC.SG', 'COM.SG', 'ABE.SG', 'PART']:
                paradigm[case] = [weak]
            else:
                paradigm[case] = [strong]
            # добавляем окончания
            if case in inflections:
                # если больше одного варианта окончания
                if len(inflections[case]) > 1:
                    stem = paradigm[case][0]
                    paradigm[case] = []
                    for inflection in inflections[case]:
                        paradigm[case].append(stem + inflection)
                else:
                    paradigm[case] = [paradigm[case][0] + inflection]
    
    if n_class == 2:
        # на долгий + краткий согласный
        strong = noun
        weak = ''.join(letters[:-2] + [letters[-2][:len(letters[-2])//2]] + [letters[-1]])
        
        for case in paradigm:
            # изменяем основу
            if case in pl + ['ACC.SG', 'GEN.SG', 'LOC.SG', 'COM.SG', 'ABE.SG', 'PART']:
                paradigm[case] = [weak]
            else:
                paradigm[case] = [strong]
            # добавляем окончания
            if case in inflections:
                if len(inflections[case]) > 1:
                    stem = paradigm[case][0]
                    paradigm[case] = []
                    for inflection in inflections[case]:
                        paradigm[case].append(stem + inflection)
                else:
                    paradigm[case] = [paradigm[case][0] + inflection]
    
    alts = noun_alt_transcript_dict[str(n_type)]['alt']
    # для первых падежей
    for case in noun_alt_transcript_dict[str(n_type)]['cases'][0]:
        
        # если в парадигме уже несколько форм
        if isinstance(paradigm[case], list):
            new_forms = []
            for form in paradigm[case]:
                for alt in alts:
                    if re.search(alt, form) != None:
                        # если есть варианты гласных
                        if isinstance(alts[alt], list):
                            # если нет вторых падежей
                            if not noun_alt_transcript_dict[str(n_type)]['cases'][1]:
                                for new_vowel in alts[alt]:
                                    form = re.sub(alt, new_vowel, form, count=1)
                                    new_forms.append(form)
                            # если есть вторые падежы
                            else:
                                form = re.sub(alt, alts[alt][0], form, count=1)
                                new_forms.append(form)
                        # если нет вариантов гласных
                        else:
                            form = re.sub(alt, alts[alt], form, count=1)
                            new_forms.append(form)
            paradigm[case] = new_forms
        # если в парадигме одна форма
        else:
            form = paradigm[case]
            for alt in alts:
                # если есть варианты гласных
                if isinstance(alts[alt], list):
                    new_forms = []
                    # если нет вторых падежей
                    if not noun_alt_transcript_dict[str(n_type)]['cases'][1]:
                        for new_vowel in alts[alt]:
                            form = re.sub(alt, new_vowel, form, count=1)
                            new_forms.append(form)
                        
                    # если есть вторые падежы
                    else:
                        form = re.sub(alt, alts[alt][0], form, count=1)
                        paradigm[case] = form
                # если нет вариантов гласных
                else:
                    form = re.sub(alt, alts[alt], form, count=1)
                    new_forms.append(new_form)
            
    # для вторых падежов
    if noun_alt_transcript_dict[str(n_type)]['cases'][1]:
        n = 0
    
    return paradigm
```
